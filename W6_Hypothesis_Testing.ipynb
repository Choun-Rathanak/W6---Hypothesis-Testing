```python


# Full code for Group1: Salary Difference Hypothesis Test
import pandas as pd
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

# Load the salary data
df = pd.read_csv('Salary_Data.csv')

# Display the first few rows
display(df.head())

# Check column names and group distribution
print('Columns:', df.columns.tolist())
print('Group counts:', df['Group'].value_counts())

# Visualize salary distribution by group
sns.boxplot(x='Group', y='Salary', data=df)
plt.title('Salary Distribution by Group')
plt.show()

# Separate the groups
group_a = df[df['Group'] == 'A']['Salary']
group_b = df[df['Group'] == 'B']['Salary']

# Descriptive statistics
print('Group A stats:', group_a.describe())
print('Group B stats:', group_b.describe())

# Perform independent t-test
t_stat, p_value = ttest_ind(group_a, group_b)

print('T-statistic:', t_stat)
print('P-value:', p_value)

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print('Reject the null hypothesis: There is a significant difference in average salary.')
else:
    print('Fail to reject the null hypothesis: No significant difference in average salary.')

# Exercise 1
# Group1: Test whether there is a significant difference in the average salary between two groups (e.g., Group A and Group B).

# Group2: Testing Whether Events (like"flu") Significantly Impact Staff Morale
# Group3: weekend has an effect on the number of electric cycles rented (cnt)
# Group4: Test whether there is a significant difference in the grdp_capita of province A(ACEH) and province B(BALI)
# Group5: Does High Stress Lead to Lower Performance?
# Identify H_o and H_1









```
