```python


# Full code for Group1: Salary Difference Hypothesis Test
import pandas as pd
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

# Load the salary data
df = pd.read_csv('Salary_Data.csv')

# Display the first few rows
display(df.head())




# Group1: Test whether there is a significant difference in the average salary between two groups (e.g., Group A and Group B).
male_salary = df[df["Gender"] == "Male"]["Salary"].dropna()
female_salary = df[df["Gender"] == "Female"]["Salary"].dropna()
t_stat, p_value = ttest_ind(male_salary, female_salary)
print(f"T-statistic: {t_stat:2f}")
print(f"P-value:{p_value:2e}")
if p_value < 0.05:
    print("There is a significant difference in average salary.")
else:
    print("There is no significant difference in average salary.")

#After test i can result There is a significant difference in average salary.



# Group2: Testing Whether Events (like"flu") Significantly Impact Staff Morale
# Group3: weekend has an effect on the number of electric cycles rented (cnt)
# Group4: Test whether there is a significant difference in the grdp_capita of province A(ACEH) and province B(BALI)
# Group5: Does High Stress Lead to Lower Performance?
# Identify H_o and H_1


# Exercise 1
# Group1: Test whether there is a significant difference in the average salary between two groups (e.g., Group A and Group B).

# Group2: Testing Whether Events (like"flu") Significantly Impact Staff Morale
# Group3: weekend has an effect on the number of electric cycles rented (cnt)
# Group4: Test whether there is a significant difference in the grdp_capita of province A(ACEH) and province B(BALI)
# Group5: Does High Stress Lead to Lower Performance?
# Identify H_o and H_1









```

```python
# Exercise 2
# Data cleaning

# Check for missing values in the dataset
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)

# Drop rows with missing Salary or Gender values 
df_cleaned = df.dropna(subset=['Salary', 'Gender'])

# reset index after cleaning
df_cleaned.reset_index(drop=True, inplace=True)

# Display cleaned dataframe info
print("Cleaned DataFrame shape:", df_cleaned.shape)
display(df_cleaned.head())
```

```python
# Exercise 3
# EDA for column testing 

# Check basic info and summary statistics for Salary and Gender columns
print(df_cleaned[["Gender", "Salary"]].info())
print(df_cleaned[["Gender", "Salary"]].describe())

# Visualize salary distribution by gender
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x="Gender", y="Salary", data=df_cleaned)
plt.title("Salary Distribution by Gender")
plt.show()
```

```python
# Exercise 4
# Discribtive statistics of testing column (count, mean, std)

print(df_cleaned.groupby("Gender")["Salary"].agg(["count", "mean", "std"]))
```

```python
# Exercise 5
# Calculate T-Statistic and P-Value
# significance level is 0.05

# Get salary data for each gender
male_salary = df_cleaned[df_cleaned["Gender"] == "Male"]["Salary"]
female_salary = df_cleaned[df_cleaned["Gender"] == "Female"]["Salary"]

# Perform independent t-test
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(male_salary, female_salary)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Check significance
if p_value < 0.05:
    print("There is a significant difference in average salary.")
else:
    print("There is no significant difference in average salary.")
```
